**Graficos de distribucion**

Histograma

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].hist(df['Ingreso Promedio'], bins=10, edgecolor='black')
axes[0].set_title('Average Income')
axes[0].set_xlabel('Income')
axes[0].set_ylabel('Frequency')

axes[1].hist(df['Ingreso familiar'], bins=10, edgecolor='black')
axes[1].set_title('Family Income')
axes[1].set_xlabel('Income')
axes[1].set_ylabel('Frequency')

axes[2].hist(df['N.I.famila'], bins=10, edgecolor='black')
axes[2].set_title('Number of Family Members')
axes[2].set_xlabel('Number of Members')
axes[2].set_ylabel('Frequency')

plt.show()
```

Diagrama de barras

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].bar(df['Genero'].value_counts().index, df['Genero'].value_counts().values)
axes[0].set_title('Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Count')

axes[1].bar(df['Tipo Trabajo'].value_counts().index, df['Tipo Trabajo'].value_counts().values)
axes[1].set_title('Work Type')
axes[1].set_xlabel('Work Type')
axes[1].set_ylabel('Count')

axes[2].bar(df['Estado Economico'].value_counts().index, df['Estado Economico'].value_counts().values)
axes[2].set_title('Economic Status')
axes[2].set_xlabel('Economic Status')
axes[2].set_ylabel('Count')

plt.show()
```

**Grafico De Tendencia**

Gráfico de líneas

ingreso promedio en funcion del genero

```python
grouped_data = df.groupby('Genero')['Ingreso Promedio'].mean().reset_index()

plt.plot(grouped_data['Genero'], grouped_data['Ingreso Promedio'])

plt.xlabel('Gender')
plt.ylabel('Average Income')
plt.title('Average Income by Gender')

plt.show()
```

ingreso promedio en funcion del tpo de trabajo

```python
grouped_data = df.groupby('Tipo Trabajo')['Ingreso Promedio'].mean().reset_index()

plt.plot(grouped_data['Tipo Trabajo'], grouped_data['Ingreso Promedio'])

plt.xlabel('Work Type')
plt.ylabel('Average Income')
plt.title('Average Income by Work Type')

plt.show()
```

ingreso familiar en función del género

```python
grouped_data = df.groupby('Genero')['Ingreso familiar'].mean().reset_index()

plt.plot(grouped_data['Genero'], grouped_data['Ingreso familiar'])

plt.xlabel('Gender')
plt.ylabel('Average Family Income')
plt.title('Average Family Income by Gender')

plt.show()
```

ingreso familiar en función del tipo de trabajo

```python
grouped_data = df.groupby('Tipo Trabajo')['Ingreso familiar'].mean().reset_index()

plt.plot(grouped_data['Tipo Trabajo'], grouped_data['Ingreso familiar'])

plt.xlabel('Work Type')
plt.ylabel('Average Family Income')
plt.title('Average Family Income by Work Type')

plt.show()
```
