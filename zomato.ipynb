**Pendograma de las variables**

```python
from scipy.cluster.hierarchy import dendrogram, linkage
data = {'Date': pd.date_range(start='2022-01-01', periods=10),
        'Open': np.random.rand(10),
        'High': np.random.rand(10),
        'Low': np.random.rand(10),
        'Close': np.random.rand(10),
        'Adj Close': np.random.rand(10),
        'Volume': np.random.rand(10)}

df = pd.DataFrame(data)
```

```python
numeric_variables = df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
linked = linkage(numeric_variables, 'single')
dendrogram(linked, orientation='top', labels=df['Date'].dt.strftime('%Y-%m-%d').tolist(), distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrograma Jerárquico de Variables Numéricas')
plt.xlabel('Índices de Datos')
plt.ylabel('Distancia')
plt.show()
```

**Analisis de componentes principales**

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

numeric_variables = df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_variables)
```

```python
pca = PCA()
principal_components = pca.fit_transform(scaled_data)
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
plt.plot(cumulative_explained_variance)
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulada')
plt.title('Análisis de Componentes Principales (PCA)')
plt.show()
```

**Análisis de Conglomerados**

```python
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(scaled_data)
df['Cluster'] = clusters
plt.scatter(df['Open'], df['Adj Close'], c=clusters, cmap='viridis')
plt.title('Análisis de Conglomerados (K-Means)')
plt.xlabel('Open')
plt.ylabel('Adj Close')
plt.show()
```

**Coeficiente de correlación de Pearson**

```python
from scipy.stats import pearsonr, spearmanr, kendalltau
correlation_coefficient, p_value = pearsonr(df['Open'], df['Adj Close'])

print(f"Coeficiente de correlación de Pearson: {correlation_coefficient}")
print(f"P-valor: {p_value}")
```

**Coeficiente de correlación de Spearman**

```python
spearman_corr, spearman_p_value = spearmanr(df['Open'], df['Adj Close'])
print(f"Coeficiente de correlación de Spearman: {spearman_corr}")
print(f"P-valor: {spearman_p_value}")
```

**Coeficiente de correlación de Kendall**

```python
kendall_corr, kendall_p_value = kendalltau(df['Open'], df['Adj Close'])
print(f"Coeficiente de correlación de Kendall: {kendall_corr}")
print(f"P-valor: {kendall_p_value}")
```
